```python
df = fdr.DataReader(samsung_code, '2010-01-01')
df.head()
```

```python
# 삼성전자의 '2022-01-01':'2022-12-31' 데이터를 df2로 지정
df1 = df.loc['2022-01-01':'2022-12-31']
# 일별 수익률
df2 = (df1['Close'].pct_change()) * 100

r = df2.dropna()
r.info()
```

```python
positive = 0
n = len(r)
for i in range(n):
  if r[i] > 0:
    positive += 1
positive
```

```python
negative = n - positive
negative
```

```python
positive_rate = positive / n
positive_rate
```

```python
import math
# 이항분포 공식

a = math.factorial(5)
b = math.factorial(3) * math.factorial(2)
number = a / b

c = (positive_rate ** 3) * (1-positive_rate) ** 2 # 성공 확률(positive_rate)이 3번 발생하고, 실패확률이 2번 발생하는 확률

prob = number * c
prob
```
